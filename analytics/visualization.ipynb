```python
gender_count=tips['sex'].value_counts()
px.pie(gender_count,gender_count.index,gender_count.values,hole=0.9,title='Gender distribution in data',
       template='plotly_dark',
       )
```

```python
def get_pie_chart(df,column,hole=0.5,template='plotly'):
    count = df[column].value_counts()
    fig = px.pie(count,
             count.index,
             count.values,
             hole=hole,
             title=f'{column} Distribution in data,'
         )
    return fig
    
```

```python
get_pie_chart(tips,'sex')
```
